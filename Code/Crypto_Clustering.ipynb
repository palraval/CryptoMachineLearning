```python
# Imports required libraries and dependencies

import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Loads the data into a Pandas DataFrame

df_market_data = pd.read_csv(
    "../Resources/crypto_market_data.csv",
    index_col="coin_id")

# Displays sample data

df_market_data.head(10)
```

```python
# Generates summary statistics

df_market_data.describe()
```

```python
# Plots data to see what's in the DataFrame

df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

scaler = StandardScaler()

scaler.fit(df_market_data)

normalized_data = scaler.transform(df_market_data)
```

```python
# Creates a DataFrame with the scaled data

scaled_df = pd.DataFrame(normalized_data)


# Copies the crypto names from the original data

cryto_names = df_market_data.index

# Sets the coinid column as index for new DataFrame

scaled_df.index = cryto_names 


# Sets the columns of new DataFrame to the columns of the original DataFrame 

scaled_df.columns = df_market_data.columns


# Displays the scaled DataFrame

scaled_df
```

---

### Find the Best Value for k Using the Original Data.

```python
# Creates a list with the number of k-values from 1 to 11

k_values = list(range(1,12))

k_values
```

```python
# Creates an empty list to store the inertia values

inertia = []

# Creates a for loop to compute the inertia with each possible value of k
for each_k in k_values:
# Inside the loop:
    # 1. Creates a KMeans model using the loop counter for the n_clusters

    model = KMeans(n_clusters= each_k, random_state= 1)
    # 2. Fits the model to the data using `scaled_df`

    model.fit(scaled_df)
    # 3. Appends the model.inertia_ to the inertia list

    inertia.append(model.inertia_)
```

```python
# Creates a dictionary with the data to plot the Elbow curve

elbow_data = {'k': k_values, 'inertia': inertia}


# Creates a DataFrame with the data to plot the Elbow curve

elbow_df = pd.DataFrame(elbow_data)

elbow_df
```

```python
# Plots a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k.

no_optimization_plot = elbow_df.hvplot.line(
                            x = "k",
                            y = "inertia",
                            title = "Elbow Curve",
                            xticks = "k"
                        )

no_optimization_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** <font color = 'aquamarine'> The best value of k seems to be 4. Any k-value after 4 results in a stagnated inertia range, so 4 clusters looks like it should be the highest number of clusters that should be accepted for this data. </font>

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initializes the K-Means model using the best value for k

model = KMeans(n_clusters= 4, random_state= 1)
```

```python
# Fits the K-Means model using the scaled data

model.fit(scaled_df)
```

```python
# Predicts the clusters to group the cryptocurrencies using the scaled data

cluster_prediction = model.predict(scaled_df)

# Prints the resulting array of cluster values.

print(cluster_prediction)
```

```python
# Creates a copy of the DataFrame

copied_scaled_df = scaled_df.copy()

copied_scaled_df
```

```python
# Adds a new column to the DataFrame with the predicted clusters

copied_scaled_df['predicted'] = cluster_prediction

# Displays sample data

copied_scaled_df
```

```python

# Creates a scatter plot using hvPlot by setting x = "price_change_percentage_24h" and y = "price_change_percentage_7d" 

no_optimization_scatter = copied_scaled_df.hvplot.scatter(
                                x = "price_change_percentage_24h",
                                y = "price_change_percentage_7d",
# Colors the graph points with the labels found using K-Means
                                by = "predicted",
# Adds the crypto name in the `hover_cols` parameter to identify the cryptocurrency represented by each data point
                                hover_cols = "coin_id"
                            )

no_optimization_scatter
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Creates a PCA model instance and sets `n_components = 3`

pca = PCA(n_components= 3)
```

```python
# Uses the PCA model with `fit_transform` to reduce to three principal components

component_data = pca.fit_transform(scaled_df)

# Views the first five rows  

component_data[:5]

```

```python
# Retrieves the explained variance to determine how much information can be attributed to each principal component

variance_ratio = pca.explained_variance_ratio_

variance_sum = variance_ratio.sum()

print()
print(f"The explained variance is {variance_ratio}")
print()
print(f"The total explained variance is {variance_sum}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** <font color = "aquamarine"> The total explained variance of the three principal components is 0.895 </font>

```python
# Creates a new DataFrame with the PCA data

pca_df = pd.DataFrame(component_data, columns= ["PCA1", "PCA2", "PCA3"])

# Copies the crypto names from the original data

original_cryto_names = df_market_data.index

# Sets the coinid column as index of new DataFrame

pca_df.index = original_cryto_names

# Displays the new DataFrame

pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11

k_list = list(range(1,11))

print(k_list)
```

```python
# Creates an empty list to store the inertia values

inertia_list = []

# Creates a for loop to compute the inertia with each possible value of k
for k in k_list:
    # Inside the loop:
    # 1. Creates a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters= k)
    # 2. Fits the model to the data using `pca_df`
    kmeans_model.fit(pca_df)
    # 3. Appends the model.inertia_ to the inertia list
    inertia_list.append(kmeans_model.inertia_)
```

```python
# Creates a dictionary with the data to plot the Elbow curve

pca_dictionary = {'k': k_list, 'inertia': inertia_list}

# Creates a DataFrame with the data to plot the Elbow curve

pca_elbow_df = pd.DataFrame(pca_dictionary)

pca_elbow_df
```

```python
# Plots a line chart with all the inertia values computed with the different values of k to visually identify the optimal value for k

optimization_plot = pca_elbow_df.hvplot.line(
                            x = 'k',
                            y = 'inertia',
                            title =  'Elbow Curve Using PCA',
                            xticks = 'k'
                        )

optimization_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** <font color = 'aquamarine'> The best value for k when using the PCA data is 4 </font>

<br>

* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:**  <font color = 'aquamarine'> No, the k-value for this method was 4 just like it was for the original data. So, it does not seem to differ. </font>

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initializes the K-Means model using the best value for k

optimal_k_model  = KMeans(n_clusters= 4)
```

```python
# Fits the K-Means model using the PCA data

optimal_k_model.fit(pca_df)
```

```python
# Predicts the clusters to group the cryptocurrencies using the PCA data

clusters = optimal_k_model.predict(pca_df)

# Prints the resulting array of cluster values

print(clusters)
```

```python
# Creates a copy of the DataFrame with the PCA data

copied_pca_df = pca_df.copy()

# Adds a new column to the DataFrame with the predicted clusters

copied_pca_df['predictions'] = clusters

# Displays the DataFrame

copied_pca_df
```

```python
# Creates a scatter plot using hvPlot by setting x="PCA1" and y="PCA2"

optimization_scatter = copied_pca_df.hvplot.scatter(
                                    x="PCA1",
                                    y="PCA2",
# Colors the graph points with the labels found using K-Means
                                    by = "predictions",
# Adds the crypto name in the `hover_cols` parameter to identify the cryptocurrency represented by each data point
                                    hover_cols = "coin_id"
                                )

optimization_scatter 
```
